```python
def single_line_handle(d):
    frm = d.split()[0]
    amnt = int(d.split()[1])
    t1 = d.split()[2]
    t2 = []
    for i in t1.split(','):
        if i in list(groups.keys()):
            t2.extend(groups[i])
        else:
            t2.append(i)
    t2 = set(t2)
    return frm,amnt,list(t2)
```

```python
def payee_handler(payee,members):
    TFs = []
    for i in members:
        if i in payee:
            TFs.append(True)
        else:
            TFs.append(False)
    return TFs
```

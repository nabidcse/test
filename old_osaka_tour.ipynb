```python
def single_row_action(single,names):
    names[single['who_paid']] = names[single['who_paid']] + single['amount']
    for_whom = single[['ali','anika', 'tamim', 'shaan', 'rayhan_bhai','mohammad', 'mou', 'mim', 'sezan_bhai']]
    for_whom = [i for i in for_whom if i is not np.nan]
    for name in for_whom:
        names[name] = names[name] - single['amount']/len(for_whom)
    return names
```

```python
names = {
    'ali' : 0,
    'anika' : 0,
    'tamim': 0,
    'shaan' :0,
    'rayhan_bhai':0,
    'mohammad':0,
    'mou':0,
    'mim':0,
    'sezan_bhai':0}

for row_i in range(len(df)):
    single = df.iloc[row_i,:]
    
    print('#############\n')
    print(single)
    names=single_row_action(single,names)
    
    display(pd.DataFrame(index=names.keys(),data=names.values(),columns=[['amount']]))
    
    print('\n')
```
